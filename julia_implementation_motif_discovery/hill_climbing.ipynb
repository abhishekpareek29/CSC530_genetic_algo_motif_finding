```julia
include("common.jl")
maximum_restarts = 10
```

```julia
function calculate_minimum_hamming_distance(motif, string)
    string_motifs = get_nth_substrings(input_motif_length, string)
    error = input_motif_length
    for substring in string_motifs
        substring_error = calculate_hamming_distance(substring, motif)
        if substring_error < error
            error = substring_error
        end
    end
    return error
end
```

```julia
function calculate_fitness_score(motif)
    fitness_score = 0
    valid = true
    for string in input_strings
        string_score = calculate_minimum_hamming_distance(motif, string)
        if string_score > input_hamming_distance
            valid = false
        end
        fitness_score += string_score
    end
    return (fitness_score, valid)
end
```

```julia
function check_motif_fitness_score_and_validation(motif)
    motif_is_valid_for_all_strings_inputs = true
    (this_motif_fitness_score, is_motif_valid) = calculate_fitness_score(motif)
    next_motif = motif
    for i in 1:maximum_restarts
        print("motif: ",motif,"\n")
        adjecent_motif = generate_adjecent_motif(motif)
        print("adjecent_motif: ",adjecent_motif,"\n")
        (adjecent_motif_fitness_score, adjecent_valid) = calculate_fitness_score(adjecent_motif)
        print("Adjecent motif fitness score: ", adjecent_motif_fitness_score, "\n")
        if (adjecent_motif_fitness_score < this_motif_fitness_score)
            print(adjecent_motif, " is better than ", next_motif, " with ", string(this_motif_fitness_score - adjecent_motif_fitness_score), " scores.\n")
            next_motif = adjecent_motif
            break
        end
    end
    if (next_motif == motif)
        return (motif, this_motif_fitness_score, is_motif_valid)
    else
        print("Going to adjecent motif.\n\n")
        return check_motif_fitness_score_and_validation(next_motif)
    end
end
```

```julia
function start_hill_climbing()
    answers = []
    for i in 1:maximum_restarts
        (motif, fitness, is_valid) = check_motif_fitness_score_and_validation(generate_random_potential_motif())
        if is_valid == true
            if isequal(motif in answers, false)
                print("\n")
                push!(answers,motif)
                print(motif, " is a discovered motif with score of ", map_fitness_to_number(fitness), "\n")
            end
        else
            print(motif, " is not a motif because of ", map_fitness_to_number(fitness), " fitness score.\n")
        end
    end
    
    if length(answers) > 0
        print(string(length(answers)), " motif(s) with length of ", string(input_motif_length), " motifs were discovered.\n")
    else
        print("No motifs with length of ", string(input_motif_length), " were found.\n")
    end
end
```

```julia
function map_fitness_to_number(fitness)
    return string(trunc(Int, ((input_motif_length * length(input_strings)) - (fitness / (length(input_strings) + 1)))))
end
```
